```python
import os
import csv
import pandas as pd
import numpy as np
import librosa
from matplotlib import pyplot as plt
import librosa.display
```

```python

#Creating a class to make a csv file for ML Model.
#Enter every detail in string.
#kindly change this location according to your location of files in line 50

class csv_work():
    
    def __init__(self,location,source_file,slice_file_name,name_of_csv_file,class_name):
        self.location = location
        self.slice_file_name = slice_file_name
        self.name_of_csv_file = name_of_csv_file
        self.source_file = source_file
        self.class_name = class_name
        
#defining a function to create csv------------------------------------------------------------------------------------------

    def csv_creation(self):
        folder_list = [self.location]
        for i in folder_list:
            my_files = [[f.name for f in os.scandir(i)if f.name.endswith('.wav')]]
            
            
#extracting out source file names-------------------------------------------------------------------------------------------         
        df = pd.DataFrame(my_files)
        df = df.T
        df.index = np.arange(1, len(df) + 1) # assign index from 1 
        df.columns = ['Source file name']     #entering column name
            
        slice_file_name_list = []
        for i in list(range(len(df))):
            slice_file_name_assign = self.slice_file_name + f'{i}'
            slice_file_name_list.append(slice_file_name_assign)


        df_slice_file_name = pd.DataFrame(slice_file_name_list)
        df_slice_file_name.index = np.arange(1, len(df_slice_file_name) + 1) # assign index from 1 
        df_slice_file_name.columns = ['slice_file_name']     #entering column name
        #print(df_slice_file_name.head(10))
        
#creating FSID data frame for the audio snippet-----------------------------------------------------------------------------
#FSID: FreesoundID of the recording where the excerpt is taken from

        FSID_list = []
        for i in list(range(len(df))):
            FSID_name = self.slice_file_name
            FSID_list.append(FSID_name)

        df_FSID_list = pd.DataFrame(FSID_list)
        df_FSID_list.index = np.arange(1, len(df_FSID_list) + 1) # assign index from 1 
        df_FSID_list.columns = ['FSID']     #entering column name

#Finding out the maximum and minimum values of amplitude--------------------------------------------------------------------
        
        #kindly change this location according to your location of files
        
        for column in df:
            location = (self.location + '/' + df[column])
            location_list = location.tolist()
            
        ls = []
        for j in location_list:
            time_series, sampling_rate = librosa.load(j)
            time_series = ls.append(time_series)

            
        max_t = []
        for k in range(len(ls)):
            max_time_series_value = max(ls[k])
            max_t.append(max_time_series_value)
        
        min_t = []
        for k in range(len(ls)):
            max_time_series_value = min(ls[k])
            min_t.append(max_time_series_value)
        
        df_max = pd.DataFrame(data=max_t)
        df_max.index = np.arange(1, len(df) + 1)
        df_max.columns = ['Maximum_Amplitude'] 
        
        df_min = pd.DataFrame(data=min_t)
        df_min.index = np.arange(1, len(df) + 1)
        df_min.columns = ['Minimum_Amplitude'] 

        
        df_location = pd.DataFrame(data=location_list)
        df_location.index = np.arange(1, len(df) + 1)
        df_location.columns = ['Location on drive'] 
        
        
#Creating class ID for each snippet------------------------------------------------------------------------------------------
#classID:
#1 = Ball Racket Hit
#2 = Ball Ground Hit

        Class_ID_list = []
        for i in list(range(len(df))):
            Class_ID_name = '1' # 1 For Ball Racket and # 2 For Ball Grou
            Class_ID_list.append(Class_ID_name)


        df_Class_ID_list_name = pd.DataFrame(Class_ID_list)
        df_Class_ID_list_name.index = np.arange(1, len(df_Class_ID_list_name) + 1) # assign index from 1 
        df_Class_ID_list_name.columns = ['Class ID']     #entering column name
        
#Creating class Name for each snippet------------------------------------------------------------------------------------------
        
        Class_Name_list = []
        for i in list(range(len(df))):
            Class_Name = self.class_name # 1 For Ball Racket and # 2 For Ball Grou
            Class_Name_list.append(Class_Name)


        df_Class_Name_list_name = pd.DataFrame(Class_Name_list)
        df_Class_Name_list_name.index = np.arange(1, len(df_Class_Name_list_name) + 1) # assign index from 1 
        df_Class_Name_list_name.columns = ['Class Name']     #entering column name


        root_audio_file_path = self.source_file
        librosa_root_audio_data,librosa_root_sample_rate = librosa.load(root_audio_file_path)
        root_audio_amplitude = librosa_root_audio_data
        np.set_printoptions(threshold=np.inf)
        
#Finding out the Start and End Time in orignal file--------------------------------------------------------------------------------------
        for column in df:
            location = (self.location + '/' + df[column])
            location_list_time = location.tolist()
        
        lt = []
        for f in location_list_time:
            librosa_audio_data,librosa_sample_rate = librosa.load(f)
            librosa_audio_data = lt.append(librosa_audio_data)
            
        start_time_array = []
        end_time_array = []
        for t in range(len(df)):
            l_real_time = lt[t]
            start_array = np.where(root_audio_amplitude == l_real_time[0] )
            start_array = np.asarray(start_array)
            arr2 = np.array([22050])
            start_array_divide = np.divide(start_array, arr2)
            start_time_array.append(start_array_divide)
            end_array = np.where(root_audio_amplitude == l_real_time[-1] )
            end_array = np.asarray(end_array)
            arr3 = np.array([22050])
            end_array_divide = np.divide(end_array, arr3)
            end_time_array.append(end_array_divide)
            
             # intialise data of lists.
            data_start = {'Start Time in original File(Sec)':start_time_array,
                         'End Time in orignal File(Sec)': end_time_array}

            # Create DataFrame
            df_time = pd.DataFrame(data_start)
            df_time.index = np.arange(1, len(df_time) + 1)
                
            
#Concatenating data frames and saving CSV file           

        
        result = pd.concat([df,df_FSID_list,df_max, df_min,df_time,df_Class_ID_list_name,df_location,df_Class_Name_list_name], axis=1)
        result.to_csv(self.name_of_csv_file)  
        return result
```
